```python
import json
def load_taillard(path='datasets/taillard_20x5.txt'):
    with open(path) as f:
        n = int(f.readline())
        tests = []
        tests_json = []
        for _ in range(n):
            test = {}
            next(f)
            nb_jobs, nb_machines, seed, ub, lb = [int(i) for i in f.readline().strip().split(" ") if i != ""]
            test["nb_jobs"] = nb_jobs
            test["nb_machines"] = nb_machines
            test["seed"] = seed
            test["upper_bound"] = ub
            test["lower_bound"] = lb
            next(f)
            jobs = []
            for _ in range(nb_machines):
                process_times = [int(i) for i in f.readline().strip().split(" ") if i != ""]
                jobs.append(process_times)
            test["jobs"] = list(zip(*jobs))
            tests_json.append(test)
            tests.append((nb_jobs, nb_machines, seed, ub, lb, list(zip(*jobs))))
    with open('datasets/taillard.json', 'w') as f:
        json.dump(tests_json, f)
    return tests_json
taillard = load_taillard('datasets/taillard_20x5.txt')
taillard_20x10 = load_taillard('datasets/taillard_20x10.txt')
```

Fonction d'utilite pour le dessin du diagramme de gantt pour une sequence donne selon les durations des jobs sur chaque machine

```python
def plot_gantt(durations, sol_indices):
    fig, ax = plt.subplots()
    # Setting up y-axis ticks
    ax.set_yticks(range(len(durations[1])))
    ax.set_yticklabels(["M" + str(i) for i in range(len(durations[1]))])

    # # Generate colors for jobs
    num_jobs, num_machines = durations.shape

    ax.set_xlabel('Time')
    ax.set_ylabel('Machines')
    ax.set_title('Flowshop Scheduling Gantt Diagram')
    colors = plt.cm.tab10(np.linspace(0, 1, num_jobs))

    job_start_times = [0] * len(durations)
    machine_start_times = [0] * len(durations[0])


    for i in sol_indices:
        for j in range(len(durations[i])):
            start_time = max(job_start_times[i], machine_start_times[j])
            duration = durations[i][j]
            color = colors[i]
            ax.barh(j, duration, left=start_time, color=color, align='center')
            ax.text(start_time + duration / 2, j, str(i), ha='center', va='center')  # Adding label for j
            job_start_times[i] = start_time + duration  # Update starting time for the job
            machine_start_times[j] = start_time + duration

    plt.show()
```

```python
def toutes_permutations(ensemble):
    """
    Génère toutes les permutations de l'ensemble donné.

    Args:
    - ensemble : ensemble d'éléments pour lequel générer les permutations

    Returns:
    - liste des permutations de l'ensemble
    """
    # Utilisation de la fonction permutations pour générer les permutations de l'ensemble
    liste_permutations = list(permutations(ensemble))

    # Retourne la liste des permutations
    return liste_permutations
```

```python
def evaluate_sequence(sequence, processing_times):
    """
    Calcule le temps total de complétion d'une séquence de tâches sur plusieurs machines.

    Args:
    - sequence : liste représentant la séquence des tâches à exécuter
    - processing_times : matrice des temps de traitement des tâches sur chaque machine

    Returns:
    - Le temps total de complétion de la séquence de tâches sur toutes les machines
    """
    completion_times = np.zeros_like(processing_times)

    # Calculating completion times for the first machine
    completion_times[0] = np.cumsum(processing_times[sequence[0]])

    # Calculating completion times for the remaining machines
    for j in range(1, processing_times.shape[1]):
        completion_times[0][j] = completion_times[0][j-1] + processing_times[sequence[0]][j]

    for i in range(1, len(sequence)):
        # Calculating completion times for each job on each machine
        completion_times[i] = completion_times[i-1] + processing_times[sequence[i]]
        for j in range(1, processing_times.shape[1]):
            completion_times[i][j] = max(completion_times[i-1][j], completion_times[i][j-1]) + processing_times[sequence[i]][j]

    # Returning the total completion time, which is the completion time of the last job in the last machine
    return completion_times[-1][-1]
```

# Jatinder Gupta 1976

```python
import math
def JatinderHeuristic(t, K):
    N = t.shape[0]
    M = t.shape[1]
    # Calculate P(k,i,m)
    p = np.zeros((K, N, 2))
    for k in range(1,K+1):
        for i in range(1,N+1):
            for m in range(1,k+1):
                p[k-1][i-1][0] += t[i-1][m-1]
            for m in range(M+1-k,M+1):
                p[k-1][i-1][1] += t[i-1][m-1]
    # Calculate R(k,i)
    R = np.zeros((N, K))  # Transpose the shape of R
    A = np.where(p[:, :, 0] >= p[:, :, 1], 1, -1)
    min_p = np.minimum(p[:, :, 0], p[:, :, 1])

    # Adjust R and A based on the transposition
    for i in range(N):
        for j in range(K):
            R[i, j] = A[j, i] / min_p[j, i]

    orders = np.zeros((K,N))
    for k in range(K):
        # Get the indices that would sort the column k
        sorted_indices = np.argsort(R[:, k])
    
        # Store the indices in the orders matrix

        orders[k] = sorted_indices
        # convert sorted indices to integers

    min_order = []
    min_cost = math.inf
    for order in orders:
        cost = evaluate_sequence(order.astype(int), t)
        if cost < min_cost:
            min_cost = cost
            min_order = order.astype(int)
    return min_order
```

# NEH

Nous utilisons l'heuristique NEH comme solution initiale pour notre problème au lieu de l'heuristique ci-dessus car elle génère un meilleur coût pour la solution initiale en comparaison avec l'heuristique ci-dessus

```python
def neh(processing_times):
    jobs, machines = processing_times.shape
    # Calcul de la somme des temps de traitement pour chaque tâche
    sum_processing_times = processing_times.sum(axis=1)

    # Création de la liste des tâches triées par ordre décroissant de la somme des temps de traitement
    sorted_jobs = np.argsort(sum_processing_times)[::-1]

    # Initialisation de la séquence avec la première tâche
    sequence = [sorted_jobs[0]]

    # Parcours des tâches triées
    for i in range(1, jobs):
        # Initialisation de la meilleure insertion et de son coût
        best_insertion = None
        best_insertion_cost = float('inf')

        # Parcours des positions d'insertion possibles dans la séquence actuelle
        for j in range(len(sequence) + 1):
            # Création d'une nouvelle séquence avec la tâche insérée à la position j
            new_sequence = sequence[:j] + [sorted_jobs[i]] + sequence[j:]

            # Calcul du temps de complétion de la nouvelle séquence
            completion_time = evaluate_sequence(new_sequence, processing_times)

            # Mise à jour de la meilleure insertion et de son coût si la nouvelle séquence est meilleure
            if completion_time < best_insertion_cost:
                best_insertion = new_sequence
                best_insertion_cost = completion_time

        # Mise à jour de la séquence avec la meilleure insertion
        sequence = best_insertion

    # Retourne la séquence optimale
    return sequence
```

pour chaque machine on calcule:

$b_i = min_{j=i+1}^{nmachines} (C_{ij})$<br>
$Tot_i = \sum_i^{njobs} C_{ij}$<br><br>
ensuite on prend le max de ces valeurs pour chaque machine et on les somme pour obtenir la borne inferieur<br>
$LB = max_i(a_i + Tot_i + b_i)$

# Greedy NEH

L'algorithme greedy NEH (Nawaz, Enscore, Ham) est une heuristique utilisée pour résoudre les problèmes d'ordonnancement de type flow shop. Son objectif est de minimiser le temps total de complétion des tâches (makespan).

Entrée : L'algorithme prend en entrée les temps de traitement des tâches pour chaque job sur chaque machine (la matrice de traitement).

Initialisation : On commence par obtenir le nombre total de tâches n à ordonnancer.

Ordre Initial des Tâches : Les tâches sont initialement ordonnées dans un ordre particulier. Cet ordre peut être choisi de différentes manières, mais dans cet algorithme, les tâches sont initialement ordonnées en fonction de leur temps total de traitement.

Initialisation des Meilleurs Résultats : On initialise les meilleures séquences de tâches et le meilleur makespan (temps total de complétion) à une valeur infinie, car nous cherchons à les minimiser.

Boucle Principale :

L'algorithme effectue un certain nombre d'itérations défini par num_iterations.

À chaque itération, une séquence partielle de tâches est construite.

On commence par sélectionner la première tâche dans la séquence partielle, qui est la première tâche dans l'ordre initial.

Ensuite, pour chaque tâche suivante, l'algorithme génère plusieurs séquences candidates en insérant la tâche à différentes positions dans la séquence partielle.

Pour chaque séquence candidate, le temps total de complétion (makespan) est calculé.

Les séquences candidates sont triées en fonction de leur makespan.

Une certaine quantité de séquences candidates est sélectionnée pour être explorée davantage. Cela est contrôlé par num_candidates.

Une séquence candidate est choisie au hasard parmi les candidats sélectionnés.

La séquence partielle est mise à jour avec la séquence candidate sélectionnée, et le makespan correspondant est mis à jour.

Si le makespan de la séquence partielle actuelle est inférieur au meilleur makespan trouvé jusqu'à présent, les meilleures séquences et le meilleur makespan sont mis à jour.

Enfin, la première tâche de la séquence initiale est déplacée à la fin de la séquence pour commencer la prochaine itération.

Sortie : Une fois toutes les itérations terminées, l'algorithme renvoie la meilleure séquence de tâches et son makespan correspondant.

```python

def Sorting_in_Descending_ordre(processing_times):
    #  Calculate the total completion time for each job
    total_completion_time = processing_times.sum(axis=1)

    #  Sort the jobs based on their total completion times in descending order
    sorted_job_indices = np.argsort(total_completion_time)[::-1]  # Sorting in descending order

    # Convert the sorted job indices to a list
    sorted_job_indices_list = sorted_job_indices.tolist()

    #  Return the sorted job indices
    return sorted_job_indices_list
```

```python
def insertion(sequence, position, value):
    #  Create a copy of the original sequence
    new_seq = sequence[:]

    #  Insert the value at the specified position in the copied sequence
    new_seq.insert(position, value)

    #  Return the modified sequence
    return new_seq
```

```python


def greedy_neh_algorithm(processing_times, num_candidates, num_iterations):
    # Get the number of jobs
    n = len(processing_times)

    # Step 1: Order jobs in descending order of total completion time
    ordered_sequence = Sorting_in_Descending_ordre(processing_times)

    # Step 2: Initialize variables to track the best sequence and its completion time
    best_sequence = []
    best_cmax = float('inf')

    # Step 3: Iterate through the specified number of iterations
    for i in range(num_iterations):
        # Step 4: Start with the first job in the ordered sequence
        partial_sequence = [ordered_sequence[0]]

        # Step 5: Iterate through the remaining jobs
        for k in range(1, n):
            # Step 6: Generate candidate sequences by inserting jobs at different positions
            candidates = []
            for job in ordered_sequence:
                if job not in partial_sequence:
                    for i in range(k+1):
                        candidate_sequence = insertion(partial_sequence, i, job)
                        candidate_cmax = evaluate_sequence(candidate_sequence, processing_times)
                        candidates.append((candidate_sequence, candidate_cmax))

            # Step 7: Sort candidate sequences based on their completion times
            candidates.sort(key=lambda x: x[1])

            # Step 8: Select the best candidate sequence from a subset of candidates
            partial_sequence, cmax = random.choice(candidates[:num_candidates])

        # Step 9: Update the best sequence and its completion time if a better solution is found
        if cmax < best_cmax:
            best_sequence = partial_sequence
            best_cmax = cmax

        # Step 10: Rotate the ordered sequence to consider a different initial job for the next iteration
        ordered_sequence.append(ordered_sequence.pop(0))

    # Step 11: Return the best sequence and its completion time
    return best_sequence, best_cmax
```

# HAM

```python
def ham_heuristic(processing_time):
    jobs, machines = processing_time.shape
    sequence = list(range(jobs))
    # Calculating the first summation
    P1 = processing_time[:,:machines//2].sum(axis=1)
    # Calculating the second summation
    P2 = processing_time[:,machines//2:].sum(axis=1)
    # Calculating the first solution, ordered by P2 - P1
    P2_P1 = P2 - P1
    solution_1 = [job for _ , job in sorted(zip(P2_P1, sequence), reverse=True)]
    # Calculating the second solution
    positives = np.argwhere(P2_P1 >= 0).flatten()
    negatives = np.argwhere(P2_P1 < 0).flatten()
    positive_indices = [job for _ , job in sorted(zip(P1[positives], positives))]
    negative_indices = [job for _ , job in sorted(zip(P2[negatives], negatives), reverse=True)]
    positive_indices.extend(negative_indices)
    # Calculating Cmax for both solutions
    Cmax1 = evaluate_sequence(solution_1, processing_time)
    Cmax2 = evaluate_sequence(positive_indices, processing_time)
    # Returning the best solution among them
    if Cmax1 < Cmax2:
        return solution_1
    else:
        return positive_indices
```

# Palmer

```python
def palmer_heuristic(processing_times):
    jobs, machines = processing_times.shape
    slope_indices = []
    for i in range(jobs):
        processing_time_sum = np.sum(processing_times[i])
        fi = 0
        for j in range(machines):
            fi += ((2*j - machines) // 2) * processing_times[i][j] / processing_time_sum
        slope_indices.append(fi)
    order = sorted(range(jobs), key=lambda k: -slope_indices[k])
    return order
```

# CDS

```python
def johnson_method(processing_times):
    jobs, machines = processing_times.shape
    copy_processing_times = processing_times.copy()
    maximum = processing_times.max() + 1
    m1 = []
    m2 = []

    if machines != 2:
        raise Exception("Johson method only works with two machines")

    for i in range(jobs):
        minimum = copy_processing_times.min()
        position = np.where(copy_processing_times == minimum)

        if position[1][0] == 0:
            m1.append(position[0][0])
        else:
            m2.insert(0, position[0][0])

        copy_processing_times[position[0][0]] = maximum

    return m1+m2

def CDS_heuristic(processing_times):
    jobs, machines = processing_times.shape
    m = machines-1
    johnson_proc_times = np.zeros((jobs,2))
    best_cost = np.inf
    best_seq = []
    for k in range(m):
        johnson_proc_times[:,0] += processing_times[:,k]
        johnson_proc_times[:,1] += processing_times[:,-k-1]
        seq = johnson_method(johnson_proc_times)
        cost = evaluate_sequence(seq,processing_times)
        if cost < best_cost:
            best_cost = cost
            best_seq = seq
    return best_seq
```

# Gupta version 1971

```python
def gupta_heuristic(t):
    N, M = t.shape
    def f(t, i):
        # Find A
        A = 1 if t[i-1, M-1] <= t[i-1, 0] else -1
        # Find min
        min_sum = min(t[i-1, j-1] + t[i-1, j] for j in range(1, M))
        return A / min_sum 
    jobs_f = [f(t, i) for i in range(1, N+1)]
    return [idx  for idx in sorted(range(len(jobs_f)), key=lambda k: jobs_f[k])]
```

```python
def eval_lower_bound(processing_times):
    jobs, machines = processing_times.shape
    times = []
    for m in range(machines):
        past_sums = []
        next_sums = []
        for j in range(jobs):
            sum1 = processing_times[j,:m].sum()
            sum2 = processing_times[j,m+1:].sum()
            past_sums.append(sum1)
            next_sums.append(sum2)
        sum_m = processing_times[:,m].sum()
        times.append(min(past_sums) + sum_m + min(next_sums))
    return max(times)
```

```python
def test(taillard_results):
    cost = np.zeros((len(taillard_results),7))

    heuristics = {
        'NEH': neh,
        'Greedy neh': greedy_neh_algorithm,
        'GUPTA 1971': gupta_heuristic,
        'GUPTA 1976 (K=2)': JatinderHeuristic,
        'CDS': CDS_heuristic,
        'HAM': ham_heuristic,
        'PALMER': palmer_heuristic
    }

    taillard_name = []
    taillard_UB = []

    for i, taillard in enumerate(taillard_results):
        taillard_name.append(taillard['dataset'])
        taillard_UB.append(taillard['taillard_UB'])
        dataset = np.array(taillard['processing_times'])
        for j, heuristic in enumerate(heuristics):
            if heuristic == 'GUPTA 1976 (K=2)':
                order = heuristics[heuristic](dataset, K=2)
            elif heuristic == 'Greedy neh':
                order, temps = heuristics[heuristic](dataset, 4, 6)
            else:
                order = heuristics[heuristic](dataset)

            cost[i,j] = evaluate_sequence(order, dataset)


    cost_df = pd.DataFrame(cost, columns=list(heuristics.keys()), index=taillard_name)
    cost_df['UB'] = taillard_UB

    cost_df['MAX_goodness'] = np.round(((np.max(cost, axis=1) - cost_df['UB']) / cost_df['UB'])*100, 2)
    cost_df['MIN_goodness'] = np.round(((np.min(cost, axis=1) - cost_df['UB']) / cost_df['UB'])*100, 2)

    return cost_df
```

```python
def plot_charts(df):
    for index, row in df.iterrows():
        plt.figure()
        plt.title(index)
        plt.bar(df.columns[:-2], row[:-2], edgecolor='black')
        plt.axhline(y=row['UB'], color='r', linestyle='--', label='UB')
        plt.xlabel('Heuristic')
        plt.ylabel('Cost')
        plt.xticks(rotation=65)
        plt.show()
```
