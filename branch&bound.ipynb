```python
import json
def load_taillard():
    path = 'datasets/taillard.txt'
    with open(path) as f:
        n = int(f.readline())
        tests = []
        tests_json = []
        for _ in range(n):
            test = {}
            next(f)
            nb_jobs, nb_machines, seed, ub, lb = [int(i) for i in f.readline().strip().split(" ") if i != ""]
            test["nb_jobs"] = nb_jobs
            test["nb_machines"] = nb_machines
            test["seed"] = seed
            test["upper_bound"] = ub
            test["lower_bound"] = lb
            next(f)
            jobs = []
            for _ in range(nb_machines):
                process_times = [int(i) for i in f.readline().strip().split(" ") if i != ""]
                jobs.append(process_times)
            test["jobs"] = list(zip(*jobs))
            tests_json.append(test)
            tests.append((nb_jobs, nb_machines, seed, ub, lb, list(zip(*jobs))))
    with open('datasets/taillard.json', 'w') as f:
        json.dump(tests_json, f)
    return tests_json
taillard = load_taillard()
```

Fonction d'utilite pour le dessin du diagramme de gantt pour une sequence donne selon les durations des jobs sur chaque machine

```python
def plot_gantt(durations, sol_indices):
    fig, ax = plt.subplots()
    # Setting up y-axis ticks
    ax.set_yticks(range(len(durations[1])))
    ax.set_yticklabels(["M" + str(i) for i in range(len(durations[1]))])

    # # Generate colors for jobs
    num_jobs, num_machines = durations.shape

    ax.set_xlabel('Time')
    ax.set_ylabel('Machines')
    ax.set_title('Flowshop Scheduling Gantt Diagram')
    colors = plt.cm.tab10(np.linspace(0, 1, num_jobs))

    job_start_times = [0] * len(durations)
    machine_start_times = [0] * len(durations[0])


    for i in sol_indices:
        for j in range(len(durations[i])):
            start_time = max(job_start_times[i], machine_start_times[j])
            duration = durations[i][j]
            color = colors[i]
            ax.barh(j, duration, left=start_time, color=color, align='center')
            ax.text(start_time + duration / 2, j, str(i), ha='center', va='center')  # Adding label for j
            job_start_times[i] = start_time + duration  # Update starting time for the job
            machine_start_times[j] = start_time + duration

    plt.show()
```

```python
def toutes_permutations(ensemble):
    """
    Génère toutes les permutations de l'ensemble donné.

    Args:
    - ensemble : ensemble d'éléments pour lequel générer les permutations

    Returns:
    - liste des permutations de l'ensemble
    """
    # Utilisation de la fonction permutations pour générer les permutations de l'ensemble
    liste_permutations = list(permutations(ensemble))

    # Retourne la liste des permutations
    return liste_permutations
```

```python
def compute_completion_time(sequence, processing_times):
    """
    Calcule le temps total de complétion d'une séquence de tâches sur plusieurs machines.

    Args:
    - sequence : liste représentant la séquence des tâches à exécuter
    - processing_times : matrice des temps de traitement des tâches sur chaque machine

    Returns:
    - Le temps total de complétion de la séquence de tâches sur toutes les machines
    """

    # Récupération du nombre de tâches et de machines
    m = processing_times.shape[1]
    n  = len(sequence)
    # Initialisation de la matrice des temps de complétion
    completion_times = np.zeros((n, m))

    # Parcours de la séquence de tâches
    for i in range(n):
        for j in range(m):
            if i == 0 and j == 0:
                # Pour la première tâche et la première machine, le temps de complétion est le temps de traitement
                completion_times[i][j] = processing_times[sequence[i]][j]
            elif i == 0:
                # Pour la première tâche sur les machines suivantes, ajoute le temps de traitement au temps de la machine précédente
                completion_times[i][j] = completion_times[i][j - 1] + processing_times[sequence[i]][j]
            elif j == 0:
                # Pour les tâches suivantes sur la première machine, ajoute le temps de traitement au temps de la tâche précédente
                completion_times[i][j] = completion_times[i - 1][j] + processing_times[sequence[i]][j]
            else:
                # Pour les tâches suivantes sur les machines suivantes, le temps de complétion est le maximum du temps précédent sur la même machine et du temps précédent sur la même tâche sur la machine actuelle, plus le temps de traitement
                completion_times[i][j] = max(completion_times[i - 1][j], completion_times[i][j - 1]) + processing_times[sequence[i]][j]

    # Retourne le temps de complétion de la dernière tâche sur la dernière machine
    return completion_times[-1][-1]
```

```python
def force_brute(permutations, processing_times):
    """
    Recherche par force brute de la permutation optimale parmi toutes les permutations données.

    Args:
    - permutations : liste des permutations à évaluer
    - processing_times : matrice des temps de traitement des tâches sur chaque machine

    Returns:
    - tuple contenant la permutation optimale et la durée minimale de complétion
    """
    min_duree = float('inf')  # Initialisation de la durée minimale à l'infini
    permutation_optimale = []  # Initialisation de la permutation optimale

    # Parcours de toutes les permutations
    for perm in permutations:
        # Calcul de la durée de complétion pour la permutation actuelle
        duree = compute_completion_time(perm, processing_times)

        # Mise à jour de la durée minimale et de la permutation optimale si la durée actuelle est plus petite
        if duree < min_duree:
            min_duree = duree
            permutation_optimale = perm

    # Retourne la permutation optimale et la durée minimale
    return permutation_optimale, min_duree
```

# Les heuristiques
L'heuristique de Jatinder et Gutpa donne une solution approximative pour la solution optimale en général de Johson à plusieurs machines, nous l'utilisons pour estimer une solution initiale pour le FSP comme une borne inférieure
https://www.tandfonline.com/doi/pdf/10.1080/05695557008974749

```python
def JatinderHeuristic(t, K):
    N = t.shape[0]
    M = t.shape[1]
    # Calculate P(k,i,m)
    p = np.zeros((K, N, 2))
    for k in range(1,K+1):
        for i in range(1,N+1):
            for m in range(1,k+1):
                p[k-1][i-1][0] += t[i-1][m-1]
            for m in range(M+1-k,M+1):
                p[k-1][i-1][1] += t[i-1][m-1]
    # Calculate R(k,i)
    R = np.zeros((N, K))  # Transpose the shape of R
    A = np.where(p[:, :, 0] >= p[:, :, 1], 1, -1)
    min_p = np.minimum(p[:, :, 0], p[:, :, 1])

    # Adjust R and A based on the transposition
    for i in range(N):
        for j in range(K):
            R[i, j] = A[j, i] / min_p[j, i]

    orders = np.zeros((K,N))
    for k in range(K):
        # Get the indices that would sort the column k
        sorted_indices = np.argsort(R[:, k])
    
        # Store the indices in the orders matrix

        orders[k] = sorted_indices
        # convert sorted indices to integers

    return orders
```

Nous utilisons l'heuristique NEH comme solution initiale pour notre problème au lieu de l'heuristique ci-dessus car elle génère un meilleur coût pour la solution initiale en comparaison avec l'heuristique ci-dessus

```python
def neh(processing_times):
    jobs, machines = processing_times.shape
    # Calcul de la somme des temps de traitement pour chaque tâche
    sum_processing_times = processing_times.sum(axis=1)

    # Création de la liste des tâches triées par ordre décroissant de la somme des temps de traitement
    sorted_jobs = np.argsort(sum_processing_times)[::-1]

    # Initialisation de la séquence avec la première tâche
    sequence = [sorted_jobs[0]]

    # Parcours des tâches triées
    for i in range(1, jobs):
        # Initialisation de la meilleure insertion et de son coût
        best_insertion = None
        best_insertion_cost = float('inf')

        # Parcours des positions d'insertion possibles dans la séquence actuelle
        for j in range(len(sequence) + 1):
            # Création d'une nouvelle séquence avec la tâche insérée à la position j
            new_sequence = sequence[:j] + [sorted_jobs[i]] + sequence[j:]

            # Calcul du temps de complétion de la nouvelle séquence
            completion_time = compute_completion_time(new_sequence, processing_times)

            # Mise à jour de la meilleure insertion et de son coût si la nouvelle séquence est meilleure
            if completion_time < best_insertion_cost:
                best_insertion = new_sequence
                best_insertion_cost = completion_time

        # Mise à jour de la séquence avec la meilleure insertion
        sequence = best_insertion

    # Retourne la séquence optimale
    return sequence
```

Fonction d'evaluation pour la borne inferieur depuis la methode utilisée pour la Lower Bound dans l'article des tests de Taillard
pour chaque machine on calcule:
<br>
$a_i = min_{j=1}^{i-1} (C_{ij})$<br>
$b_i = min_{j=i+1}^{nmachines} (C_{ij})$<br>
$Tot_i = \sum_i^{njobs} C_{ij}$<br><br>
ensuite on prend le max de ces valeurs pour chaque machine et on les somme pour obtenir la borne inferieur<br>
$LB = max_i(a_i + Tot_i + b_i)$

```python
def eval_lower_bound(processing_times):
    jobs, machines = processing_times.shape
    times = []
    for m in range(machines):
        past_sums = []
        next_sums = []
        for j in range(jobs):
            sum1 = processing_times[j,:m].sum()
            sum2 = processing_times[j,m+1:].sum()
            past_sums.append(sum1)
            next_sums.append(sum2)
        sum_m = processing_times[:,m].sum()
        times.append(min(past_sums) + sum_m + min(next_sums))
    return max(times)
```

# Structure du noeud

```python
class Node:
    def __init__(self, jobs, remaining_jobs, parent=None, lower_bound=float('inf'), elague=False):
        """
        Initialise un nœud de l'arbre de recherche.

        Args:
        - jobs : liste des tâches affectées jusqu'à présent dans ce nœud
        - remaining_jobs : ensemble des tâches non affectées dans ce nœud
        - parent : nœud parent de ce nœud (par défaut None pour le nœud racine)
        - lower_bound : borne inférieure associée à ce nœud (par défaut infinie)
        """
        self.jobs = jobs
        self.remaining_jobs = remaining_jobs
        self.parent = parent
        self.lower_bound = lower_bound
        self.elague = elague

    def __str__(self):
        """
        Retourne une représentation sous forme de chaîne de caractères du nœud.
        """
        # return ''.join(map(str, self.jobs))
        return f"Node(jobs={self.jobs}, remaining_jobs={self.remaining_jobs}, lower_bound={self.lower_bound})"
```

# Implementation
On implemente l'algorithme de branch and bound pour le FSP avec elagage en utilisant la borne inferieur calculée par la methode ci-dessus, avec:<br>
- Chaque niveau represente le noued qui est executé sur la branche(une solution ou ordonnancement partielle)<br>
- Fonction d'evaluation pour le temps d'execution pour les jobs ordonnee au niveau du noeud courant<br>
- Calcul la borne inferieur depuis la methode utilisée pour la Lower Bound dans l'article des tests de Taillard<br>
- Un parcours DFS  

```python
def branch_and_bound(processing_times, initial_solution, initial_cost):
    """
    Implémente l'algorithme Branch and Bound pour résoudre un problème de flow shop.

    Args:
    - processing_times : matrice des temps de traitement des tâches sur chaque machine
    - initial_solution : solution initiale
    - initial_cost : coût initial de la solution initiale

    Returns:
    - tuple contenant la meilleure solution, son coût et le nombre total d'itérations effectuées
    """
    jobs, machines = processing_times.shape

    # Initialisation du nœud racine avec une liste vide de tâches et l'ensemble de tous les indices de tâches
    root_node = Node([], set(range(jobs)))

    # Initialisation de la meilleure solution et de son coût initial
    best_solution = initial_solution.copy()
    best_solution_cost = initial_cost

    # Initialisation de la liste des nœuds à explorer avec le nœud racine
    nodes = [root_node]

    # Initialisation du compteur d'itérations
    iteration_count = 0

    # Boucle principale pour explorer les nœuds de l'arbre de recherche
    while nodes:
        # Retirer le dernier nœud de la liste des nœuds à explorer
        node = nodes.pop()

        # Exploration des voisins du nœud actuel en ajoutant une nouvelle tâche à la séquence actuelle
        for job in node.remaining_jobs:
            # Création d'un nouveau nœud enfant en ajoutant la tâche
            child_jobs = node.jobs + [job]
            child_remaining_jobs = node.remaining_jobs - {job}

            # Calcul du lower bound pour le nouveau nœud enfant
            child_lower_bound = compute_completion_time(child_jobs, processing_times)
            # Si le nœud enfant est une feuille, calculer son coût et le comparer à la meilleure solution actuelle
            if not child_remaining_jobs:
                if child_lower_bound < best_solution_cost:
                    best_solution = child_jobs
                    best_solution_cost = child_lower_bound
                continue

            # Si le nœud enfant n'est pas une feuille, vérifier s'il peut améliorer la meilleure solution actuelle
            child_lower_bound += eval_lower_bound(processing_times[list(child_remaining_jobs)])
            if child_lower_bound > best_solution_cost:
                child_node = Node(child_jobs, child_remaining_jobs, parent=node, lower_bound=child_lower_bound)
                nodes.append(child_node)

        # Incrémentation du compteur d'itérations
        iteration_count += 1

        if iteration_count % 10000 == 0:
            print(f"Iteration {iteration_count} - Best Solution: {best_solution} - Best Cost: {best_solution_cost}")
    # Retour de la meilleure solution trouvée, son coût et le nombre total d'itérations
    return best_solution, best_solution_cost, iteration_count
```

Dessous on donne une version sans elagage de l'algorithme de branch and bound pour le FSP qui est une forme de recherche exhaustive exploitant la forme de l'arbre de recherche pour le FSP

```python
def branch_and_bound_pure(processing_times, initial_solution, initial_cost):
    # Détermination du nombre de tâches et de machines
    #il parcourt tous l'espace de recherche
    jobs, machines = processing_times.shape

    # Initialisation du nœud racine avec une liste vide de tâches et l'ensemble de tous les indices de tâches
    root_node = Node([], set(range(jobs)))

    # Initialisation de la meilleure solution et de son coût initial
    best_solution = initial_solution.copy()
    best_solution_cost = initial_cost

    # Initialisation de la liste des nœuds à explorer avec le nœud racine
    nodes = [root_node]

    # Initialisation du compteur d'itérations
    iteration_count = 0

    # Boucle principale pour explorer les nœuds de l'arbre de recherche
    while nodes:
        # Retirer le dernier nœud de la liste des nœuds à explorer
        node = nodes.pop()

        # Exploration des voisins du nœud actuel en ajoutant une nouvelle tâche à la séquence actuelle
        for job in node.remaining_jobs:
            # Création d'un nouveau nœud enfant en ajoutant la tâche
            child_jobs = node.jobs + [job]
            child_remaining_jobs = node.remaining_jobs - {job}

            # Si le nœud enfant est une feuille (toutes les tâches ont été assignées), calculer son coût
            if not child_remaining_jobs:
                child_lower_bound = compute_completion_time(child_jobs, processing_times)

                # Comparaison avec la meilleure solution trouvée jusqu'à présent
                if child_lower_bound < best_solution_cost:
                    best_solution = child_jobs
                    best_solution_cost = child_lower_bound
                    continue
            else:
                # Si le nœud enfant n'est pas une feuille, calculer son lower bound
                child_lower_bound = compute_completion_time(child_jobs, processing_times)

                # Création du nœud enfant et ajout à la liste des nœuds à explorer
                child_node = Node(child_jobs, child_remaining_jobs, parent=node, lower_bound=child_lower_bound)
                nodes.append(child_node)

        # Incrément du compteur d'itérations
        iteration_count += 1

    # Retour de la meilleure solution trouvée, son coût et le nombre total d'itérations
    return best_solution, best_solution_cost, iteration_count
```

# Tests
On a teste l'algorithme implimente sur trois instance aleatoires de FSP qui ont tous genere une solution optimale en un temps raisonnable pour un nombre de machines 5 et [7,8,9,10] jobs

```python
test_instances = []
np.random.seed(33334)
for njobs in range(6, 11):
    data_aleatoire = np.random.randint(low=10, high=50, size=(njobs,5))
    initial_solution = neh(data_aleatoire)
    # initial_solution = list(range(njobs))
    initial_cost = compute_completion_time(initial_solution, data_aleatoire)
    start_time = time.time()
    best_solution, best_cost, i = branch_and_bound(data_aleatoire, initial_solution, initial_cost)
    elapsed_time = time.time() - start_time

    test_instances.append({
        "dataset": 'testdata_nj_' + str(njobs),
        "processing_times": data_aleatoire.tolist(),
        "taillard_LB": None,
        "taillard_UB": None,
        "initial_solution": initial_solution,
        "initial_cost": initial_cost,
        "best_solution": best_solution,
        "best_cost": best_cost,
        "time": elapsed_time,
        "Number of nodes visited": i
    })
    print("###############################")
    print(f'Results of Branch & Bound:')
    print(f'Best sequence is {best_solution} with a makespan of {best_cost}.')
    print(f'No. Nodes visited is {i}.')
    print(f'Elapsed time of {elapsed_time} seconds.')
```
